```python
## All the olympians who have participated in the summer olympics are put into the dataframe
df_summer = df_olympic[df_olympic['Season'] == 'Summer']
# drop any duplicates occuring for any athlete from matching Name Sex and Year and drop NAs from Height
avg_data = df_summer.drop_duplicates(subset=['Name', 'Sex', 'Year'], keep='first').dropna(subset=['Height'])
avg_data
```

```python
#Plot the line plots with standard deviation handled by errorbar='sd'
fig, ax = plt.subplots(figsize=(10,8))
sns.lineplot(x = 'Year', y = 'Height', data = avg_data, hue = 'Sex', linewidth=1.5, palette=['b', 'r'], ax=ax, errorbar = 'sd')
sns.despine()
sns.set(font_scale = 1)
#Removes grid squares
ax.grid(False)
ax.set_ylim(140, 200)
handles, labels = plt.gca().get_legend_handles_labels()
#To order the legend as M and F
order = [0, 1]
plt.legend([handles[i] for i in order], [labels[i] for i in order],loc='lower left', framealpha=0)
ax.set_xlabel('Year',fontsize=15)
ax.set_ylabel('Height(cm)',fontsize=15)
ax.set_title('Average Olympian Height (+/- std dev) in the Summer Olympics',fontsize=15)
```

### Q4) The bar plot below shows the number of athletes that competed in the 2012 (dark blue) and 2016 (light blue) Olympics from a set of different countries. The Venn diagram acts as the legend â€“ it informs the meaning of each colour. Additionally, it shows the overlap on the number of athletes (from these countries) that competed in both Olympics, and how many competed in just one of these Olympics.
Steps in the form of code comments  \
References:
https://stackoverflow.com/questions/34854570/how-to-join-two-or-more-dataframes-in-pandas-python \
https://www.geeksforgeeks.org/python-pandas-melt/ \

```python
#Dropping duplicates from the original olympic data based on ID and year
df_olympic = df_olympic.drop_duplicates(subset=['ID', 'Year'])
df_olympic

#Retrieve the Participants of 2012 and 2016 olympics and group top 11 countries and rename the index as a counter
df_olympic_12_16_vals = df_olympic[(df_olympic['Year'] == 2012) | (df_olympic['Year'] == 2016)]
df_olympic_12_16 = df_olympic_12_16_vals.groupby(['Team'])['ID'].count().reset_index(name='Counter').sort_values(by=['Counter'],ascending=False)[:11]

df_olympic_12_16

#The total count of athletes for 2012 and 2016
totalcount = df_olympic_12_16.Counter.sum()

#Transferring top 10 countries to a list for its use to retrieve the counts based on the specific year and sort in descending order of players from participating countries
df_olympic_2012_2016_top_team_names = df_olympic_12_16.Team.tolist()
df_olympic_2012_2016_top_team_names

df_olympic_2012 = df_olympic[(df_olympic['Year'] == 2012) & (df_olympic['Team'].isin(df_olympic_2012_2016_top_team_names))]
df_olympic_2012 = df_olympic_2012.groupby(['Team','Year'])['ID'].count().reset_index(name='Counter').sort_values(by=['Counter'],ascending=False)

#The total count of athletes for 2012 with intersection/Duplicates
df_olympic_2012_count = df_olympic_2012.Counter.sum()

# Get only the names of countries with players in 2012 olympics and use this list to retrieve the olympian count from 2016 from these countries 
df_olympic_2012_top_team_names = df_olympic_2012.Team.tolist()
df_olympic_2012_top_team_names

df_olympic_2016 = df_olympic[df_olympic['Team'].isin(df_olympic_2012_top_team_names)]
df_olympic_2016 = df_olympic_2016[df_olympic_2016['Year'] == 2016].groupby(['Team','Year'])['ID'].count().reset_index(name='Counter').sort_values(by=['Counter'],ascending=False)
df_olympic_2016

#Setting team as a candidate key index for 2016 dataframe column and referencing the primary key Team column from 2012 dataframe to keep the order undisturbed
df_olympic_2016 = df_olympic_2016.set_index('Team')
df_olympic_2016 = df_olympic_2016.reindex(index=df_olympic_2012['Team'])
df_olympic_2016 = df_olympic_2016.reset_index()
df_olympic_2016

#The total count of athletes for 2016 With Intersection/Duplicates
df_olympic_2016_count = df_olympic_2016.Counter.sum()

# Performing inner join for 2012 and 2016 dataframes with Team, Counter and Year of both the tables based on the primary and candidate keys
olympic_sport_df = df_olympic_2012.merge(df_olympic_2016, on='Team', how='inner').reset_index()
olympic_sport_df

#Perform a melt on the data to transform to rows to plot
olympic_sport_df_refined = olympic_sport_df[['Team','Counter_x','Counter_y']]
olympic_sport_df_refined=pd.melt(olympic_sport_df_refined, id_vars =['Team'], value_vars =['Counter_x','Counter_y'])
olympic_sport_df_refined

# total Counts without duplicates

totalcount_without_duplicates = df_olympic_12_16_vals[((df_olympic_12_16_vals['Year'] == 2012)|(df_olympic_12_16_vals['Year'] == 2016)) & (df_olympic_12_16_vals['Team'].isin(df_olympic_2012_2016_top_team_names))]
totalcount_without_duplicates = len(totalcount_without_duplicates.drop_duplicates('ID'))

#Intersection count is the total count - the count without duplicates
intersection_count = totalcount - totalcount_without_duplicates

only_2012_count = totalcount - df_olympic_2016_count - intersection_count

only_2016_count = totalcount - df_olympic_2012_count - intersection_count

#Plot the bar plots

fig, ax = plt.subplots(figsize=(10,8))
sns.set(font_scale = 1)
#sns.barplot(x = 'value', y = 'Team', data = olympic_sport_df_refined, color='red', label = '2012',ax = ax)
sns1 = sns.barplot(x = 'value', y = 'Team', data = olympic_sport_df_refined, hue = 'variable', palette='tab20', ax=ax)
#Remove legend
sns1.get_legend().remove()
ax.set_xlabel('Number of Athletes',fontsize=15)
ax.set_title('Change in Number of Athletes per Country from 2012 to 2016',fontsize=15)

#Set inset axis for the venn diagram plot based on below coordinates
axis2 = sns1.inset_axes([0.7,0.1,0.3,0.3])
#Set tab20 colourmap and set color to the venn diagram 
cm = plt.cm.get_cmap('tab20')
#Plot the venn diagram
c = venn2(subsets = (only_2012_count, only_2016_count, intersection_count), set_labels = ('2012', '2016'),ax=axis2,alpha=1)
#Add the Colours individually
c.get_patch_by_id('10').set_color(cm(0))
c.get_patch_by_id('01').set_color(cm(1))
c.get_patch_by_id('11').set_color(cm(0.9))

```
