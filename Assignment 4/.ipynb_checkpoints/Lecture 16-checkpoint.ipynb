# Seaborn Example

Let's make a line plot to address the question: *What are the average number of views per month for channels with trending videos?*

We will just look at the top 5 channels (based on number of trending videos) for months in 2018.

```python
df = pd.read_csv('CAvideos.csv', parse_dates=['trending_date', 'publish_time'])
```

Get the top 5 channels based on number of (unique) videos that went trending. 

```python
df[['video_id', 'channel_title']].groupby('channel_title').nunique().reset_index().sort_values(by='video_id', ascending=False).head().channel_title
```

```python
top5_channels = df[['video_id', 'channel_title']].groupby('channel_title').nunique().reset_index().sort_values(by='video_id', ascending=False).head().channel_title
```

```python
top5_channels
```

```python
df = df[df.channel_title.isin(top5_channels)]
```

Restrict to just videos that went trending in 2018.

```python
df.info()
```

```python
df = df[df.trending_date.dt.year == 2018]
```

Remove duplicate videos. Only the most recent entry for the video is kept, which is done by sorting by date in descending order and keeping only the first entry.

```python
df = df.sort_values(by='trending_date', ascending=False).drop_duplicates(subset=['video_id'], keep='first')
```

Extract the month from the trending date information (which was parsed as datetime when the data was initially loaded).

```python
df['trending_month'] = df.trending_date.dt.month
```

### Matplotlib Implementation

First, we will create the line of average views per month using Matplotlib.

We need to get the data for the plot by using groupby to calculate the average views per month per channel.

```python
df_m = df[['channel_title', 'trending_month', 'views']].groupby(['channel_title', 'trending_month']).mean().reset_index()
```

```python
fig, ax = plt.subplots()

for channel in top5_channels:
    df_tmp = df_m[df_m.channel_title == channel]
    ax.plot(df_tmp.trending_month, df_tmp.views, label=channel)
    
ax.legend()
ax.set_xlabel('Trending Month')
ax.set_ylabel('views')
```

### Seaborn implementation

Now let's plot the same data using Seaborn's ``lineplot()`` function.

```python
sns.lineplot(x = 'trending_month', y = 'views', data = df, hue = 'channel_title', legend=False)
```

We have created the same plot using one line of code.

In fact, the plot is even better than the Matplotlib plot because it has included shading to represent the variance in number of views. 

In this case, the shading represents the 95% confidence interval for the average.

Not only did Seaborn calculate the averages for us - previously we had to manually calculate them using groupby - but it has included shading for the confidence intervals for us as well. It would have added a lot of extra work to calculate those ourselves and manually draw the shading using Matplotlib.

**Seaborn functions return a Matplotlib Axes object.** All the regular Matplotlib functionality can then be used with it.

```python
ax = sns.lineplot(x = 'trending_month', y = 'views', data = df, hue = 'channel_title')

ax.set_ylim((0, 1.5e6))
ax.set_xlabel('Trending Month')
ax.set_title('Top 5 Trending YouTube Channels')
```

**Seaborn can also draw onto existing Axes objects.** Just pass your Axes object to the function using the ``ax=`` parameter.

```python
fig, ax = plt.subplots(figsize=(7,5))

sns.lineplot(x = 'trending_month', y = 'views', data = df, hue = 'channel_title', ax=ax)
```

```python
fig
```

Note that the pandas DataFrame was directly passed to the Seaborn function call. The x and y data were specified by the 
column name of the DataFrame, same as with the column that the colour and markers were based on. Seaborn works well with tidy data.

# Seaborn Distribution Plots

Seaborn offers a variety of plot types for distributions, some of which we have previously studied and some new.

The distribution of a quantitative variable is its spread of values.

## Histograms

Histograms show one distribution at a time. This is different than box or violin plots which show multiple distributions at once.

In this example, the histogram is the distribution of views for all videos for VikatanTV.

```python
sns.histplot(x='views', data=df_vikatan)
```

By default, the histogram is the count of items in each bin. This can be changed using the ``stat=`` parameter, which takes the following values:
    
    count: show the number of observations in each bin
    frequency: show the number of observations divided by the bin width
    probability: or proportion: normalize such that bar heights sum to 1
    percent: normalize such that bar heights sum to 100
    density: normalize such that the total area of the histogram equals 1

```python
sns.histplot(x='views', data=df_vikatan, stat='percent')
```

Visually, the shape of the distribution is the same - it is the quantitative scale on the y-axis that is different.

The bin sizes have been automatically set in the preceding examples. Using the ``bins=`` parameter, there are a number of built-in ways to automatically create the bins, or the number of bins can be specified, or a list of bin edges can be used.

```python
sns.histplot(x='views', data=df_vikatan, bins=40)
```

Multiple distributions can be plotted at once using the ``hue=`` parameter. In this case, let's plot the distribution for category of videos.

```python
sns.histplot(x='views', data=df_vikatan, hue='category_id')
```

By default, the two histograms are layered on top of each other. You can specify to Seaborn how multiple distributions should be handled by the ``multiple=`` keyword, which accepts either ``layer`` (default), ``stack``, ``dodge`` or ``fill``.

```python
sns.histplot(x='views', data=df_vikatan, hue='category_id', multiple='stack')
```

## Box Plots

Box plots show the median, inter-quartile range (IQR) and extended points that are outliers to the distribution.

The box plot below shows the distribution of views for the videos corresponding to that month. Unlike histograms, which are ideal for studying a single distribution, box plots are excellent for comparing distributions to one another.

```python
fig, ax = plt.subplots(figsize=(9,6))

sns.boxplot(x='trending_month', y='views', data=df_vikatan, ax=ax)
```

We can see that February has a wide range, with a couple of videos that are abnormally high. This helps explain the large error bars that were present in the line and bar plots.

We can also add nested grouping to our box plot. Let's plot the view distribution per month, but also layer in the video categories using ``hue=``.

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.boxplot(x='trending_month', y='views', data=df_vikatan, hue='category_id', ax=ax)
```

Each month now shows two box plots - one for category id 24 and a second for id 43. 

Also notice how Seaborn has changed the colour scheme to something sensible. We will discuss Seaborn's colour options further below.

## Violin Plots

We can represent the same plots using violin plots.

First is the distribution of views per month.

```python
sns.violinplot(x='trending_month', y='views', data=df_vikatan)
```

The width of the violins can be set using ``scale=``.
- ``area``: each violin will have the same area. 
- ``count``: the width of the violins will be scaled by the number of observations in that bin. 
- ``width``: each violin will have the same width.

```python
fig, ax = plt.subplots(figsize=(16,9))
sns.violinplot(x='trending_month', y='views', data=df_vikatan, scale='count', ax=ax)
```

The interior of the violin has a miniature box plot overlaid on it.

The mini box plot can be changed to lines at the position of the quartiles instead with the ``inner=`` parameter. In effect, this combines the best elements of the box plot and violin plot together.

```python
fig, ax = plt.subplots(figsize=(16,9))
sns.violinplot(x='trending_month', y='views', data=df_vikatan, scale='count', inner='quartiles', ax=ax)
```

We can create the second box plot in the previous section also by using the ``hue=`` parameter.

```python
fig, ax = plt.subplots(figsize=(16,9))
sns.violinplot(x='trending_month', y='views', data=df_vikatan, hue='category_id', ax=ax)
```

This is much easier to create than with Matplotlib!

One option that may work well in some cases is to use ``split=True|False``. Rather than drawing the two violins side-by-side, they are drawn as two halves of a single violin.

```python
fig, ax = plt.subplots(figsize=(16,9))
sns.violinplot(x='trending_month', y='views', data=df_vikatan, hue='category_id', split=True, ax=ax)
```

Overall, violin plots created through Seaborn are actually useful and easy to work with, unlike Matplotlib.

## Colour

We have not specified any colour choices thus far, and instead have relied on Seaborn to choose colours on its own.

There are a number of colour palettes available with Seaborn. The current palette can be obtained with ``sns.color_palette()``.

```python
sns.color_palette()
```

To set a palette, the ``sns.set_palette()`` function can be used. Palettes can often be specified in plot function calls, as well, and individual colours specified, if that fine-grain customization is needed.

```python
sns.set_palette("tab10")
```

```python
palette = sns.color_palette()
```

```python
palette
```

There are many options.

```python
sns.color_palette()
```

```python
sns.set_palette("RdBu")
sns.boxplot(x='trending_month', y='views', data=df_vikatan)
```

The colour palette can also be specified in the plot function call using ``palette=``. This accepts either the name of the palette, a palette object, or a list of colours.

```python
sns.boxplot(x='trending_month', y='views', data=df_vikatan, palette='hls')
```

```python
palette = sns.color_palette('Paired')

sns.boxplot(x='trending_month', y='views', data=df_vikatan, palette=palette)
```

Instead of using a palette, a list of specific colours can be used instead. The colours can be specified by name or by using RGB values.

```python
colours = ['red', 'blue', 'green']#, 'orange', 'yellow', '#880088']

sns.boxplot(x='trending_month', y='views', data=df_vikatan, palette=colours)
```

All of the plots in this notebook accept the ``palette=`` parameter.

# Summary

Seaborn offers a high-level interface to Matplotlib. It still allows you to do all the fine-grain tuning in Matplotlib that you have learned how to do, but permits you to skip many of the common, tedious tasks. It also creates plots with a nice aesthetic, and can do many transformations for you. Most functions use the same syntax, with ``x=``, ``y=``, ``data=``, ``hue=`` and ``palette=``.

There are a number of built-in options to customize the visual look of your plot. And, if needed, they can be further customized using Matplotlib.

The documentation for Seaborn's API is excellent: https://seaborn.pydata.org/api.html
