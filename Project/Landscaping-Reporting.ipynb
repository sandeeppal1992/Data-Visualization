```python
df_landscaping
df_profits = df_landscaping

##Profits earned
df_profits['material_costs'] = df_profits['material_costs'].astype(float)
df_profits['profits'] = df_landscaping['invoice_amount'] - df_landscaping['material_costs']

## Number of days effort days and the total days prohect was held
df_profits['start_date'] = pd.to_datetime(df_profits['start_date'], errors='coerce')
df_profits['completion_date'] = pd.to_datetime(df_profits['completion_date'], errors='coerce')
df_profits['request_date'] = pd.to_datetime(df_profits['completion_date'], errors='coerce')

#type(df_profits.request_date)
df_profits['effort_spent'] = df_landscaping['completion_date'] - df_landscaping['request_date']
#type(df_profits.effort_spent)
df_profits['effort_spent'] = df_profits['effort_spent'] + pd.to_timedelta(1, unit='D')
df_profits['total_days'] = df_landscaping['completion_date'] - df_landscaping['start_date']
df_profits['total_days'] = df_profits['total_days'] + pd.to_timedelta(1, unit='D')

df_profits['year'] = pd.DatetimeIndex(df_profits['completion_date']).year
df_profits['month'] = pd.DatetimeIndex(df_profits['completion_date']).month

# Work with profits
#Profits by job type
profits_by_job_type_sum = df_profits[['job_type','profits','customer_postal_code','customer_satisfaction','customer_id','year','month']].groupby(['job_type','month']).sum().reset_index().sort_values(by=['profits'],ascending=False)
profits_by_job_type_mean = df_profits[['job_type','profits','customer_postal_code','customer_satisfaction','customer_id','year','month']].groupby(['job_type','month']).mean().reset_index().sort_values(by=['profits'],ascending=False)
#profits_by_job_type_sum = profits_by_job_type_sum.fillna(0)
#profits_by_job_type_mean = profits_by_job_type_mean.fillna(0)

#Profits by a particular customer
profits_by_customer_sum = df_profits[['job_type','profits','customer_id']].groupby(['job_type','customer_id']).sum().reset_index().sort_values(by=['profits'],ascending=False)[:10]
profits_by_customer_mean = df_profits[['job_type','profits','customer_id']].groupby(['job_type','customer_id']).mean().reset_index().sort_values(by=['profits'],ascending=False)[:10]

#Profits per postal code
profits_by_postal_sum = df_profits[['job_type','profits','customer_postal_code']].groupby(['job_type','customer_postal_code']).sum().reset_index().sort_values(by=['profits'],ascending=False)
profits_by_postal_mean = df_profits[['job_type','profits','customer_postal_code']].groupby(['job_type','customer_postal_code']).mean().reset_index().sort_values(by=['profits'],ascending=False)


#Profits by satisfied and unsatisfied customers
df_profits_satisfied = df_profits[df_profits['customer_satisfaction'] > 5]
df_profits_unsatisfied = df_profits[df_profits['customer_satisfaction'] <= 5]

profits_by_satisfied_customer_sum = df_profits_satisfied[['job_type','profits','customer_postal_code','customer_satisfaction','customer_id']].groupby(['job_type','customer_satisfaction']).sum().reset_index().sort_values(by=['profits'],ascending=False)
profits_by_unsatisfied_customer_sum = df_profits_unsatisfied[['job_type','profits','customer_postal_code','customer_satisfaction','customer_id']].groupby(['job_type','customer_satisfaction']).sum().reset_index().sort_values(by=['profits'],ascending=False)

#profits_by_unsatisfied_customer_sum
#profits_by_satisfied_customer_sum
#profits_by_customer_mean
#profits_by_customer_sum
#df_mostrecent[['comments_disabled', 'likes', 'dislikes', 'like_ratio']].groupby('comments_disabled').mean().reset_index()
#profits_by_job_type_sum
#profits_by_job_type_sum
#df_profits

#profits_by_satisfied_customer_sum
profits_by_unsatisfied_customer_sum
profits_by_unsatisfied_customer_sum = profits_by_unsatisfied_customer_sum.fillna(0)



#Load the example flights dataset and convert to long-form

heat_profits_by_job_type_sum = profits_by_job_type_sum.pivot(index='job_type', columns='month', values='profits')
heat_profits_by_job_type_sum = heat_profits_by_job_type_sum.fillna(0)

# # Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 9))
#sns.heatmap(heat, annot=True, fmt="d", linewidths=.5, ax=ax)
sns.heatmap(data=heat_profits_by_job_type_sum, ax = ax,annot=True,fmt="0.1f")
plt.title("Heat Map representing total profits across job types spanning different months")
print(heat_profits_by_job_type_sum)


heat_profits_by_job_type_mean = profits_by_job_type_mean.pivot(index='job_type', columns='month', values='profits')
heat_profits_by_job_type_mean = heat_profits_by_job_type_mean.fillna(0)

# # Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 9))
#sns.heatmap(heat, annot=True, fmt="d", linewidths=.5, ax=ax)
sns.heatmap(data=heat_profits_by_job_type_mean, ax = ax,annot=True,fmt="0.1f")
plt.title("Heat Map representing average profits/job across job types spanning different months")


heat_profits_by_customer_sum = profits_by_customer_sum.pivot(index='job_type', columns='customer_id', values='profits')
heat_profits_by_customer_sum = heat_profits_by_customer_sum.fillna(0)

# # Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 9))
#sns.heatmap(heat, annot=True, fmt="d", linewidths=.5, ax=ax)
sns.heatmap(data=heat_profits_by_customer_sum, ax = ax,annot=True,fmt="0.1f")
plt.title("Heat Map representing total profits across job types spanning different customers")

heat_profits_by_customer_mean = profits_by_customer_mean.pivot(index='job_type', columns='customer_id', values='profits')
heat_profits_by_customer_mean = heat_profits_by_customer_mean.fillna(0)

# # Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 9))
#sns.heatmap(heat, annot=True, fmt="d", linewidths=.5, ax=ax)
sns.heatmap(data=heat_profits_by_customer_mean, ax = ax,annot=True,fmt="0.1f")
plt.title("Heat Map representing total profits across job types spanning different customers")

heat_profits_by_postal_sum = profits_by_postal_sum.pivot(index='job_type', columns='customer_postal_code', values='profits')
heat_profits_by_postal_sum = heat_profits_by_postal_sum.fillna(0)

# # Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 9))
#sns.heatmap(heat, annot=True, fmt="d", linewidths=.5, ax=ax)
sns.heatmap(data=heat_profits_by_postal_sum, ax = ax,annot=True,fmt="0.1f",cmap = 'mako')
plt.title("Heat Map representing total profits across job types spanning different customers")

heat_profits_by_postal_mean = profits_by_postal_mean.pivot(index='job_type', columns='customer_postal_code', values='profits')
heat_profits_by_postal_mean = heat_profits_by_postal_mean.fillna(0)
# # Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 9))
#sns.heatmap(heat, annot=True, fmt="d", linewidths=.5, ax=ax)
sns.heatmap(data=heat_profits_by_postal_mean, ax = ax,annot=True,fmt="0.1f",cmap = 'mako')
plt.title("Heat Map representing total profits across job types spanning different postal codes")

heat_profits_by_satisfied_customer_sum = profits_by_satisfied_customer_sum.pivot(index='job_type', columns='customer_satisfaction', values='profits')
heat_profits_by_satisfied_customer_sum = heat_profits_by_satisfied_customer_sum.fillna(0)
# # Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 9))
#sns.heatmap(heat, annot=True, fmt="d", linewidths=.5, ax=ax)
sns.heatmap(data=heat_profits_by_satisfied_customer_sum, ax = ax,annot=True,fmt="0.1f",cmap=sns.cubehelix_palette(as_cmap=True))
plt.title("Heat Map representing total profits across job types spanning different customer satisfaction")

heat_profits_by_unsatisfied_customer_sum = profits_by_unsatisfied_customer_sum.pivot(index='job_type', columns='customer_satisfaction', values='profits')
heat_profits_by_unsatisfied_customer_sum = heat_profits_by_unsatisfied_customer_sum.fillna(0)
# # Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 9))
#sns.heatmap(heat, annot=True, fmt="d", linewidths=.5, ax=ax)
sns.heatmap(data=heat_profits_by_unsatisfied_customer_sum, ax = ax,annot=True,fmt="0.1f",cmap=sns.cubehelix_palette(as_cmap=True))
plt.title("Heat Map representing total profits across job types spanning different customer satisfaction")




# df_diamonds1 = df_diamonds[['carat', 'colour', 'clarity']].groupby(['colour', 'clarity']).mean().reset_index()
# df_diamonds1 = df_diamonds1.pivot(index='colour', columns='clarity', values='carat')
# df_diamonds1 = df_diamonds1.reindex(['SI2', 'SI1', 'VS2', 'VS1','VVS2','VVS1','IF'], axis="columns")
# fig, ax = plt.subplots(figsize = (9,6))
# sns.heatmap(data=df_diamonds1, ax = ax,cmap = 'mako')




# fig, ax = plt.subplots(figsize=(12,10))
 
# # declaring exploding pie
# #explode = [0, 0.1, 0, 0, 0]
# # define Seaborn color palette to use
# palette_color = sns.color_palette('pastel')
  
# # plotting data on chart
# #plt.pie(profits_by_customer_sum, labels=job_type, colors=palette_color,explode=explode, autopct='%.0f%%')
# plt.pie(profits_by_job_type_sum, labels=profits_by_job_type_sum.profits, colors=palette_color)

  
# # displaying chart
# plt.show()
```
