## Vizualization 1 - Horizontal Bar Plot

### How can we find which job type is not performing well and has a low satisfaction level among it's clientbase?

Explanatory Visualization \
The below visualization describes the Average customer satisfaction for each work type. From the bar chart we can see tasks like water feature and retaining wall have very less customer satisfaction levels; these areas need an improvement to improve the company reputation. \
Have represented each of them into 3 classification classes based on the performance. \ 
Ratings above 7 are classified into good category and below 6 needs a performance review.

```python
cust_satisfaction=df_landscaping[['job_type', 'customer_satisfaction']].groupby('job_type').mean()
cust_satisfaction_sorted = cust_satisfaction.sort_values(by='customer_satisfaction', ascending=False).reset_index()
cust_satisfaction_sorted
fig = plt.Figure()
ax = fig.add_axes([0, 0, 1, 1])
fig, ax = plt.subplots()
cust_satisfaction_sorted['color'] = ['red' if (x < 6) else 'green' if x > 7 else 'orange' for x in cust_satisfaction_sorted.customer_satisfaction]
# set up color palette
sns.barplot(x=cust_satisfaction_sorted.customer_satisfaction, y=cust_satisfaction_sorted.job_type, orient='h',palette=cust_satisfaction_sorted['color'])
#handles = [plt.Rectangle((0,0),1,1, color=color) for color in set(cust_satisfaction_sorted['color'])]
handles = [plt.Rectangle((0,0),1,1, color=color) for color in ['Green','Orange','Red']]
labels = ['Good Performance', 'Meets Standards','Needs Improvement']
plt.legend(handles, labels,fontsize="small")
for i, v in enumerate(cust_satisfaction_sorted.customer_satisfaction):
    plt.text(v + 0.1, i - 0.25, str(round(v, 2)), color='black')
ax.set_xlim(0,9)
ax.set_title(label = 'Customer Satisfaction average for each work type')
ax.invert_yaxis()
```

## Vizualization 2 - Box Plot

### Which are the work areas where the top 5 paid employees with hourly wage > 22$ perform good and where they lack?

Exploratory Visualization \
The above box plot visualization shows the performance metrics of its highly paid employees grouped categorically with wages > $22/hour for every work item. \
● This will help the company to explore the skills where each employee is lacking and the tasks where the employee can see a scope for improvement. For ex: the low paid employee performs better than the highest paid employee in tree pruning by observing their medians in the box plots.\
● This metrics will help the company decide which tasks need to be allocated to which employee to maximize the profits and improve the company
reputation. This data is extracted from the Calendar and employee datasets.

```python
import warnings
warnings.filterwarnings('ignore')
calendar = df_calendar
df_employees_top = df_employees[df_employees['hourly_wage'] >= 22].sort_values(by='hourly_wage', ascending=False)
employee = df_employees_top.employee_id.values.tolist()
a4_dims = (15, 12)
appended_df  = pd.DataFrame() # Initilize
df_user_test = pd.DataFrame()
for i in range(len(employee)):
    calendar[i] = df_calendar[employee[i]]
    test_j_ids = list(set(calendar[i].values.tolist()))
    df_user_test = df_landscaping[df_landscaping['job_id'].isin(test_j_ids)]
    df_user_test["eid"] = employee[i]
    appended_df = pd.concat([appended_df, df_user_test], ignore_index=True)
sns1=sns.catplot(data=appended_df, x="job_type", y="customer_satisfaction", hue="eid", kind="box",ax=ax,sharex=True,palette="pastel",height=8, aspect=2,
                 notch=False,linewidth=1,width = 0.6,showmeans=True,medianprops={'color': 'red', 'lw': 2},showcaps=True)

medians = appended_df.groupby(['job_type'])['customer_satisfaction'].median().values
nobs = appended_df['job_type'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]
 
# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    ax.text(pos[tick],
            medians[tick] + 0.03,
            nobs[tick],
            horizontalalignment='center',
            size='x-small',
            color='w',
            weight='semibold')
ax.set(xlabel="Job Type", ylabel="Customer Satisfaction")
ax.set_xticklabels('Number of Athletes',fontsize=15,rotation=90)
ax.grid(True, axis='both')

plt.title("Customer Satisfaction Rating for the top 5 paid Workers")
```

## Vizualization 3 - Interactive Heatmap

### How much are the profits earned by the company based on Job type, top paying customers,postal code and satisfied/unsatisfied customer?

Exploratory Visualization \
Shown above are few interactive scenarios of the single heatmap visualization representing the total/average profit margins across different job types for different parameters. \
Some of them are: 
1) Heat Map representing total/Average profits across job types spanning different months. This will help company explore the trends of months when business is best 
2) Heat Map representing total/Average profits across job types spanning top 10 paying customers. This will give insights on the spending habits of the clients. 
3) Heat Map representing total/Average profits across job types spanning different postal codes. This will give insights on the spending habits of the clients from different areas. 
4) Heat Map representing total profits across job types among satisfied/unsatisfied customers. This will give insights on whether satisfaction impacts profit margins anyhow. 

Above are the 8 scenarios that can be handled by the single interactive plot. 

```python
options = ['Total', 'Average']
def f(yaxis,x):

    
    df_landscaping = pd.read_csv('landscaping.csv',na_filter=False)
    df_profits = df_landscaping

    ##Profits earned
    df_profits['material_costs'] = df_profits['material_costs'].astype(float)
    df_profits['profits'] = df_landscaping['invoice_amount'] - df_landscaping['material_costs']

    ## Number of days effort days and the total days prohect was held
    df_profits['start_date'] = pd.to_datetime(df_profits['start_date'], errors='coerce')
    df_profits['completion_date'] = pd.to_datetime(df_profits['completion_date'], errors='coerce')
    df_profits['request_date'] = pd.to_datetime(df_profits['completion_date'], errors='coerce')

    df_profits['effort_spent'] = df_landscaping['completion_date'] - df_landscaping['request_date']
    df_profits['effort_spent'] = df_profits['effort_spent'] + pd.to_timedelta(1, unit='D')
    df_profits['total_days'] = df_landscaping['completion_date'] - df_landscaping['start_date']
    df_profits['total_days'] = df_profits['total_days'] + pd.to_timedelta(1, unit='D')

    df_profits['year'] = pd.DatetimeIndex(df_profits['completion_date']).year
    df_profits['month'] = pd.DatetimeIndex(df_profits['completion_date']).month

# Work with profits
#Profits by job type
    profits_by_job_type_sum = df_profits[['job_type','profits','customer_postal_code','customer_satisfaction','customer_id','year','month']].groupby(['job_type','month']).sum().reset_index().sort_values(by=['profits'],ascending=False)
    profits_by_job_type_mean = df_profits[['job_type','profits','customer_postal_code','customer_satisfaction','customer_id','year','month']].groupby(['job_type','month']).mean().reset_index().sort_values(by=['profits'],ascending=False)


#Profits by a particular customer
    profits_by_customer_sum = df_profits[['job_type','profits','customer_id']].groupby(['job_type','customer_id']).sum().reset_index().sort_values(by=['profits'],ascending=False)[:10]
    profits_by_customer_mean = df_profits[['job_type','profits','customer_id']].groupby(['job_type','customer_id']).mean().reset_index().sort_values(by=['profits'],ascending=False)[:10]

#Profits per postal code
    profits_by_postal_sum = df_profits[['job_type','profits','customer_postal_code']].groupby(['job_type','customer_postal_code']).sum().reset_index().sort_values(by=['profits'],ascending=False)
    profits_by_postal_mean = df_profits[['job_type','profits','customer_postal_code']].groupby(['job_type','customer_postal_code']).mean().reset_index().sort_values(by=['profits'],ascending=False)


#Profits by satisfied and unsatisfied customers
    df_profits_satisfied = df_profits[df_profits['customer_satisfaction'] > 5]
    df_profits_unsatisfied = df_profits[df_profits['customer_satisfaction'] <= 5]

    profits_by_satisfied_customer_sum = df_profits_satisfied[['job_type','profits','customer_postal_code','customer_satisfaction','customer_id']].groupby(['job_type','customer_satisfaction']).sum().reset_index().sort_values(by=['profits'],ascending=False)
    profits_by_unsatisfied_customer_sum = df_profits_unsatisfied[['job_type','profits','customer_postal_code','customer_satisfaction','customer_id']].groupby(['job_type','customer_satisfaction']).sum().reset_index().sort_values(by=['profits'],ascending=False)

#profits_by_unsatisfied_customer_sum
#profits_by_satisfied_customer_sum
#profits_by_customer_mean
#profits_by_customer_sum
#df_mostrecent[['comments_disabled', 'likes', 'dislikes', 'like_ratio']].groupby('comments_disabled').mean().reset_index()
#profits_by_job_type_sum
#profits_by_job_type_sum
#df_profits
    profits_by_unsatisfied_customer_sum
    profits_by_unsatisfied_customer_sum = profits_by_unsatisfied_customer_sum.fillna(0)



#Load the example flights dataset and convert to long-form


    if ((yaxis == 'Profits by job type') and (x == 'Total')):
        heat_profits_by_job_type_sum = profits_by_job_type_sum.pivot(index='job_type', columns='month', values='profits')
        heat_profits_by_job_type_sum = heat_profits_by_job_type_sum.fillna(0)
        ## Draw a heatmap with the numeric values in each cell
        f, ax = plt.subplots(figsize=(9, 8))
        sns.heatmap(data=heat_profits_by_job_type_sum, ax = ax,annot=True,fmt="0.1f")
        plt.title("Heat Map representing total profits across job types spanning different months")

    elif ((yaxis == 'Profits by job type') and (x == 'Average')):
        heat_profits_by_job_type_mean = profits_by_job_type_mean.pivot(index='job_type', columns='month', values='profits')
        heat_profits_by_job_type_mean = heat_profits_by_job_type_mean.fillna(0)
    # # Draw a heatmap with the numeric values in each cell
        f, ax = plt.subplots(figsize=(9, 8))
        sns.heatmap(data=heat_profits_by_job_type_mean, ax = ax,annot=True,fmt="0.1f")
        plt.title("Heat Map representing average profits/job across job types spanning different months")

    
    elif ((yaxis == 'Profits by customer ID') and (x == 'Total')):
        heat_profits_by_customer_sum = profits_by_customer_sum.pivot(index='job_type', columns='customer_id', values='profits')
        heat_profits_by_customer_sum = heat_profits_by_customer_sum.fillna(0)
    # # Draw a heatmap with the numeric values in each cell
        f, ax = plt.subplots(figsize=(9, 8))
        sns.heatmap(data=heat_profits_by_customer_sum, ax = ax,annot=True,fmt="0.1f")
        plt.title("Heat Map representing total profits across job types spanning different top 10 paying customers")

    elif ((yaxis == 'Profits by customer ID') and (x == 'Average')):
       
        heat_profits_by_customer_mean = profits_by_customer_mean.pivot(index='job_type', columns='customer_id', values='profits')
        heat_profits_by_customer_mean = heat_profits_by_customer_mean.fillna(0)
    # # Draw a heatmap with the numeric values in each cell
        f, ax = plt.subplots(figsize=(9, 8))
        sns.heatmap(data=heat_profits_by_customer_mean, ax = ax,annot=True,fmt="0.1f")
        plt.title("Heat Map representing average profits across job types spanning different top 10 paying customers")

    elif ((yaxis == 'Profits by postal code') and (x == 'Total')):

        heat_profits_by_postal_sum = profits_by_postal_sum.pivot(index='job_type', columns='customer_postal_code', values='profits')
        heat_profits_by_postal_sum = heat_profits_by_postal_sum.fillna(0)
        # # Draw a heatmap with the numeric values in each cell
        f, ax = plt.subplots(figsize=(9, 8))
        sns.heatmap(data=heat_profits_by_postal_sum, ax = ax,annot=True,fmt="0.1f",cmap = 'mako')
        plt.title("Heat Map representing total profits across job types spanning different Postal codes")

    elif ((yaxis == 'Profits by postal code') and (x == 'Average')):
        
        heat_profits_by_postal_mean = profits_by_postal_mean.pivot(index='job_type', columns='customer_postal_code', values='profits')
        heat_profits_by_postal_mean = heat_profits_by_postal_mean.fillna(0)
    # # Draw a heatmap with the numeric values in each cell
        f, ax = plt.subplots(figsize=(9, 8))
        sns.heatmap(data=heat_profits_by_postal_mean, ax = ax,annot=True,fmt="0.1f",cmap = 'mako')
        plt.title("Heat Map representing total average across job types spanning different postal codes")

    elif ((yaxis == 'Profits by customer satisfaction') and (x == 'Total')):    
        heat_profits_by_satisfied_customer_sum = profits_by_satisfied_customer_sum.pivot(index='job_type', columns='customer_satisfaction', values='profits')
        heat_profits_by_satisfied_customer_sum = heat_profits_by_satisfied_customer_sum.fillna(0)
    # # Draw a heatmap with the numeric values in each cell
        f, ax = plt.subplots(figsize=(9, 8))
        sns.heatmap(data=heat_profits_by_satisfied_customer_sum, ax = ax,annot=True,fmt="0.1f",cmap=sns.cubehelix_palette(as_cmap=True))
        plt.title("Heat Map representing total profits across job types spanning customer satisfaction > 5")

    elif ((yaxis == 'Profits by customer satisfaction') and (x == 'Average')):    
        heat_profits_by_unsatisfied_customer_sum = profits_by_unsatisfied_customer_sum.pivot(index='job_type', columns='customer_satisfaction', values='profits')
        heat_profits_by_unsatisfied_customer_sum = heat_profits_by_unsatisfied_customer_sum.fillna(0)
    # # Draw a heatmap with the numeric values in each cell
        f, ax = plt.subplots(figsize=(9, 8))
        sns.heatmap(data=heat_profits_by_unsatisfied_customer_sum, ax = ax,annot=True,fmt="0.1f",cmap=sns.cubehelix_palette(as_cmap=True))
        plt.title("Heat Map representing total profits across job types spanning customer satisfaction < 5")

    
interact(f, yaxis = ['Profits by job type', 'Profits by customer ID', 'Profits by postal code', 'Profits by customer satisfaction'] , x=widgets.ToggleButtons(options=options,description=" "));
```

## Vizualization 4 - Swarmplot with derived datatypes

### Which is the time taken to perform a job for the customer? How can it ve improved

Exploratory Visualization: \
The below swarmplot displays the total time/effort taken in days spent by the employees to execute jobs from a particular jobtype across different months. \
This will help the company to find out any discrepancy in any work taking place and if anything seems out of the ordinary.

```python
df_landscaping
df_duration = df_landscaping

## Number of days effort days and the total days prohect was held
df_duration['start_date'] = pd.to_datetime(df_duration['start_date'], errors='coerce')
df_duration['completion_date'] = pd.to_datetime(df_duration['completion_date'], errors='coerce')
df_duration['request_date'] = pd.to_datetime(df_duration['completion_date'], errors='coerce')

#type(df_profits.request_date)
df_duration['effort_spent'] = df_landscaping['completion_date'] - df_landscaping['request_date']
#type(df_profits.effort_spent)
df_duration['effort_spent'] = df_duration['effort_spent'] + pd.to_timedelta(1, unit='D')
df_duration['total_days_per_job'] = df_landscaping['completion_date'] - df_landscaping['start_date']
df_duration['total_days_per_job'] = df_duration['total_days_per_job'] + pd.to_timedelta(1, unit='D')

df_duration['year'] = pd.DatetimeIndex(df_duration['completion_date']).year
df_duration['month'] = pd.DatetimeIndex(df_duration['completion_date']).month
#print(df_duration)

duration_by_month = df_duration[['job_type','total_days_per_job','month']].groupby(['job_type','month']).sum().reset_index().sort_values(by=['total_days_per_job'],ascending=False)
duration_by_month

#Convert month to datetime
duration_by_month.month = pd.to_datetime(duration_by_month.month, format='%m')
#Convert to month in names
duration_by_month.month = duration_by_month.month.dt.month_name().str.slice()

job_count_by_month = df_landscaping[['job_type','month','total_days_per_job']].groupby(['job_type','month']).count().rename(columns={'total_days_per_job': 'job_count'})
job_count_by_month
#sns.set_theme(style="whitegrid", palette="pastel")

ax=sns.catplot(
    data=duration_by_month, x="total_days_per_job", y="month", hue="job_type",native_scale=False, zorder=1,size=13,ax=ax,height=6, aspect=1.75,kind='swarm',palette="Paired"
)
for ax in ax.axes.flat:
    ax.grid(True, axis='both')
plt.xticks(rotation = 90)
#sns.set_style("darkgrid")

sns.despine(bottom = True, left = True)
plt.title("Swarm plot representing total time spent in days for jobs types during different months of 2022")
#job_count_by_month
#duration_by_month
```
